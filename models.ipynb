###**One-Hot Encoding**###

```
dummy_toss=pd.get_dummies(df1['TossDecision'])
df_enc=pd.concat([df1,dummy_toss],axis='columns')
df_enc.drop(columns=['TossDecision','bat','year'],inplace=True) #the bat column does not provide any new information therefore dropped
#df_enc
dummy_method=pd.get_dummies(df1['method'])
df_enc=pd.concat([df_enc,dummy_method],axis=1)
df_enc.drop(columns=['method','Normal'],inplace=True)  #The normal column does not add any new value so dropping it along with the original column
df_enc.head()
```

*CREATING* A NEW COLUMN WITH THE VALUE AS ID+BATTING TEAM WHICH WILL ACT AS THE ID FOR AN INNING

```

df_enc['id_inning']=df_enc['ID'].astype(str)+df_enc['BattingTeam']
df_enc.sample(10)
```

```
list=df_enc['id_inning'].unique()
len(list)

# list=df_enc['Date'].unique()
# len(list)
```

```
#Total run column in the df_enc is only giving us the runs scored on that particular ball, so to get the runs scored in an innings we need
# to first group each innings column and then sum the total run..

numeric_columns = df_enc.select_dtypes(include=['number']).columns.tolist()
numeric_columns.append('id_inning')  # Include the grouping column
# print(numeric_columns)
# Perform groupby and sum only on numeric columns
df_grouped = df_enc[numeric_columns].groupby(by=['id_inning']).sum()
# print(df_grouped)
# Extract the 'total_run' column and reset the index
df_copy = df_grouped[['total_run']].reset_index()

# Display the first few rows
df_copy.head()
# df_copy.sample(10)
```

```
#Now we are going to merge the total runs scored in the initial in our initial df_enc

data=pd.merge(left=df_enc,right=df_copy,on='id_inning',how='right')
data.head(6)
```

```
# Now we are going to covert each ball as overs dot ball number e.g 1.2, 2.5,..

data['overs']=data['overs']+(data['ballnumber'])/10
data.drop(columns=['ballnumber'],inplace=True)
data.head(12)
```

###**Adding runs and wickets at every instant of the match**###

```
# Now we going to add runs and wickets at every instant of the match

# Create an empty DataFrame
data_final = pd.DataFrame()
# Here list is basically an array of the id_inning representing each innings
for id in list:
    inning_data = data[data['id_inning'] == id]

    # Calculate cumulative sums
    cum_total = inning_data['total_run_x'].cumsum()
    cum_wicket = inning_data['isWicketDelivery'].cumsum()

    # Add cumulative sums to DataFrame
    inning_data['runs'] = cum_total
    inning_data['wickets'] = cum_wicket

    # Use pd.concat to append the DataFrame
    data_final = pd.concat([data_final, inning_data], ignore_index=True)

data_final.head(20)
```

###**Custom Accuracy Model**###

```
def custom_accu(y_test,y_pred,thresh):

  correct=0
  length=len(y_pred)
  for i in range(0,length):

    if(abs(y_test[i]-y_pred[i])<=thresh):

      correct=correct+1
  return((correct/length)*100)
```

###**SVM Model**###

```
from sklearn.svm import SVR
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.model_selection import GridSearchCV, cross_val_score, train_test_split
from sklearn.metrics import mean_squared_error

# Assuming X and y are already defined and preprocessed
# Reshape y to be a 1D array
y = y.ravel()

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=1)

# Define column transformer for preprocessing
numeric_features = X.select_dtypes(include=['int64', 'float64']).columns
categorical_features = X.select_dtypes(include=['object']).columns

numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])

# Define the parameter grid for SVR
param_grid = {
    'svr__C': [0.1, 1, 10],
    'svr__epsilon': [0.1, 0.2, 0.5],
    'svr__kernel': ['linear', 'rbf', 'poly']
}

# Create a pipeline with preprocessing and SVR
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('svr', SVR())
])

# Setup GridSearchCV with the pipeline
grid_search = GridSearchCV(estimator=pipeline, param_grid=param_grid, cv=3, n_jobs=-1, verbose=2)
grid_search.fit(X_train, y_train.ravel())

# Best parameters and model
best_svr = grid_search.best_estimator_

# Predict with the best model
y_predict = best_svr.predict(X_test)

# Calculate mean squared error and custom accuracy
mse = mean_squared_error(y_test, y_predict)
custom_accuracy = custom_accu(y_test, y_predict, 10)

print(f'Best parameters: {grid_search.best_params_}')
print(f'Custom accuracy is: {custom_accuracy}%')
print(f'Mean Squared Error: {mse}')

# Optional: Perform cross-validation on the full data
cv_scores = cross_val_score(best_svr, X, y.ravel(), cv=3, scoring='neg_mean_squared_error')
print(f'Cross-validated MSE: {-np.mean(cv_scores)}')
```

###**Linear Regression with Ridge and Lasso Regularization**###

```
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, PolynomialFeatures, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.impute import SimpleImputer

# Assuming X and y are already defined and preprocessed
# Reshape y to be a 1D array
y = y.ravel()

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=1)

# Define column transformer for preprocessing
numeric_features = X.select_dtypes(include=['int64', 'float64']).columns
categorical_features = X.select_dtypes(include=['object']).columns

numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])

# Create pipelines for Linear Regression, Ridge, and Lasso
linear_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('poly', PolynomialFeatures(degree=2, include_bias=False)),
    ('regressor', LinearRegression())
])

ridge_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('poly', PolynomialFeatures(degree=2, include_bias=False)),
    ('regressor', Ridge(alpha=1.0))
])

lasso_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('poly', PolynomialFeatures(degree=2, include_bias=False)),
    ('regressor', Lasso(alpha=0.1))
])

# Fit and evaluate the Linear Regression model
linear_pipeline.fit(X_train, y_train)
y_pred = linear_pipeline.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
custom_accuracy = custom_accu(y_test, y_pred, 10)

print('Linear Regression:')
print(f'Mean Absolute Error: {mae}')
print(f'Mean Squared Error: {mse}')
print(f'R² Score: {r2}')
print(f'Custom accuracy is: {custom_accuracy}%')

# Perform cross-validation for Linear Regression
cv_scores = cross_val_score(linear_pipeline, X, y, cv=5, scoring='neg_mean_squared_error')
print(f'Linear Regression Cross-validated MSE: {-np.mean(cv_scores)}')

# Fit and evaluate the Ridge model
ridge_pipeline.fit(X_train, y_train)
y_pred_ridge = ridge_pipeline.predict(X_test)

ridge_mae = mean_absolute_error(y_test, y_pred_ridge)
ridge_mse = mean_squared_error(y_test, y_pred_ridge)
ridge_r2 = r2_score(y_test, y_pred_ridge)
ridge_custom_accuracy = custom_accu(y_test, y_pred_ridge, 12)

print('\nRidge Regression:')
print(f'Mean Absolute Error: {ridge_mae}')
print(f'Mean Squared Error: {ridge_mse}')
print(f'R² Score: {ridge_r2}')
print(f'Custom accuracy is: {ridge_custom_accuracy}%')

# Perform cross-validation for Ridge Regression
cv_scores_ridge = cross_val_score(ridge_pipeline, X, y, cv=5, scoring='neg_mean_squared_error')
print(f'Ridge Regression Cross-validated MSE: {-np.mean(cv_scores_ridge)}')

# Fit and evaluate the Lasso model
lasso_pipeline.fit(X_train, y_train)
y_pred_lasso = lasso_pipeline.predict(X_test)

lasso_mae = mean_absolute_error(y_test, y_pred_lasso)
lasso_mse = mean_squared_error(y_test, y_pred_lasso)
lasso_r2 = r2_score(y_test, y_pred_lasso)
lasso_custom_accuracy = custom_accu(y_test, y_pred_lasso, 12)

print('\nLasso Regression:')
print(f'Mean Absolute Error: {lasso_mae}')
print(f'Mean Squared Error: {lasso_mse}')
print(f'R² Score: {lasso_r2}')
print(f'Lasso Custom accuracy is: {lasso_custom_accuracy}%')

# Perform cross-validation for Lasso Regression
cv_scores_lasso = cross_val_score(lasso_pipeline, X, y, cv=5, scoring='neg_mean_squared_error')
print(f'Lasso Regression Cross-validated MSE: {-np.mean(cv_scores_lasso)}')
```

###**Random Forest Regressor**###

```
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.impute import SimpleImputer
import numpy as np

# Assuming X and y are already defined and preprocessed
# Reshape y to be a 1D array
y = y.ravel()

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=1)

# Define column transformer for preprocessing
numeric_features = X.select_dtypes(include=['int64', 'float64']).columns
categorical_features = X.select_dtypes(include=['object']).columns

numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])

# Define the RandomForestRegressor with GridSearchCV
rfr = RandomForestRegressor(random_state=1)
param_grid = {
    'regressor__n_estimators': [100, 200, 300],
    'regressor__max_depth': [None, 10, 20, 30],
    'regressor__min_samples_split': [2, 5, 10],
    'regressor__min_samples_leaf': [1, 2, 4]
}

# Create preprocessing and training pipeline
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', rfr)
])

# Setup GridSearchCV with the pipeline
grid_search = GridSearchCV(estimator=pipeline, param_grid=param_grid, cv=3, n_jobs=2, verbose=2)
grid_search.fit(X_train, y_train)

# Best parameters and model
best_model = grid_search.best_estimator_

# Predict with the best model
y_pred = best_model.predict(X_test)

# Evaluate the model
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
custom_accuracy = custom_accu(y_test, y_pred, 10)

print(f'Best parameters: {grid_search.best_params_}')
print(f'Mean Absolute Error: {mae}')
print(f'Mean Squared Error: {mse}')
print(f'R² Score: {r2}')
print(f'Custom accuracy is: {custom_accuracy}%')

# Optional: Perform cross-validation on the full data
cv_scores = cross_val_score(best_model, X, y, cv=3, scoring='neg_mean_squared_error')
print(f'Cross-validated MSE: {-np.mean(cv_scores)}')
```
